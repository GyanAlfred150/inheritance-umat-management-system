# UMaT Management System

```
class Person:
    def __init__(self,name,age):
        self.name=name; self.age=age
    def display_info(self):
        print(f'Name: {self.name}\nAge: {self.age}')
class Student(Person):
    def __init__(self,name,age,student_id):
        super().__init__(name,age); self.student_id=student_id
    def enroll_course(self,course): print(f'{self.name} enrolled in {course}')
    def display_info(self):
        super().display_info(); print('Student ID:',self.student_id)
class Lecturer(Person):
    def __init__(self,name,age,employee_id):
        super().__init__(name,age); self.employee_id=employee_id
    def teach_course(self,course): print(f'{self.name} teaches {course}')
    def display_info(self):
        super().display_info(); print('Employee ID:',self.employee_id)
s=Student('Alice',20,'ST001'); l=Lecturer('Dr Bob',45,'EMP101')
s.display_info(); s.enroll_course('Python')
l.display_info(); l.teach_course('OOP')
```
